### Attributes that look like they have a good linear relationship

It looks like attributes 5 and 6 have a good linear relationship

```python
sns.regplot(data=df, x=df[6], y=df[5])
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

X = df[[6]]
y = df[[5]]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

```python
lin_model = LinearRegression()
lin_model.fit(X_train,y_train)
```

```python
y_pred = lin_model.predict(X_test)
```

```python
print(lin_model.intercept_)
```

```python
from sklearn import metrics
print('R^2:',metrics.r2_score(y_test, y_pred))
```

```python
from yellowbrick.regressor import ResidualsPlot
res = ResidualsPlot(lin_model)
res.fit(X_train,y_train)
res.score(X_test,y_test)
res.show()
```
